```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mixture_proportions = [0.3, 0.5, 0.2]
mean = [1.2, 4.3, 7.7]
std_dev = [0.34, 0.69, 1.4]
```

```python
start = min(mean) - 3 * max(std_dev)
end = max(mean) + 3 * max(std_dev)
x = np.linspace(start, end, 500)

y0 = stats.norm.pdf(x, mean[0], std_dev[0])
y1 = stats.norm.pdf(x, mean[1], std_dev[1])
y2 = stats.norm.pdf(x, mean[2], std_dev[2])
y = y0 + y1 + y2

plt.plot(x, y0, x, y1, x, y2)
plt.plot(x, y)
plt.show()
```

```python
def generate_data():
    data = []
    for _ in range(200):
        num = np.random.choice([0,1,2], p=mixture_proportions)
        data.append(stats.norm(loc=mean[num], scale=std_dev[num]).rvs())
    return data

def Expectation(data, mean, std_dev, weight):
    posteriori = []
    for d in data:
        prob = []
        sum = 0
        for i in range(3):
            sum += weight[i] * stats.norm.pdf(d, mean[i], std_dev[i])
            print('SUM', sum)
        for i in range(3):
            prob.append(weight[i] * stats.norm.pdf(d, mean[i], std_dev[i]) / sum)
        posteriori.append(prob)
    return posteriori

def Maximisation(data, posteriori):
    weight = [p/len(data) for p in
                      [sum(i) for i in zip(*posteriori)]]
    temp1 = [[i * x for i in prob] for prob, x in zip(posteriori, data)]
    temp1 = [sum(i) for i in zip(*temp1)]
    mean = [a/(b*len(data)) for a, b in zip(temp1, weight)]
    
    temp2 = [[i * (x**2) for i in prob] for prob, x in zip(posteriori, data)]
    temp2 = [sum(i) for i in zip(*temp2)]
    temp2 = [a/(b*len(data)) for a, b in zip(temp2, weight)]
    std_dev = [a - (b**2) for a, b in zip(temp2, mean)]
    return mean, std_dev, weight
```

```python
data = generate_data()
random_mean = [1, 2, 3]
random_dev = [1, 1, 1]
random_weight = [0.2, 0.6, 0.2]

for _ in range(1):
    posteriori = Expectation(data, random_mean, random_dev, random_weight)
    random_mean, random_dev, random_weight = Maximisation(data, posteriori)
    
```
